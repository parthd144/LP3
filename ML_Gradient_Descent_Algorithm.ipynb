```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and its derivative
def f(x):
    return (x + 3) ** 2

def f_derivative(x):
    return 2 * (x + 3)
```

```python
# Gradient Descent parameters
learning_rate = 0.1  # Step size
max_iter = 100       # Maximum number of iterations
tolerance = 1e-6     # Convergence tolerance
```

```python
# Starting point
x = 2
x_values = [x]  # To keep track of x values for plotting
```

```python
# Gradient Descent Loop
for i in range(max_iter):
    gradient = f_derivative(x)
    new_x = x - learning_rate * gradient
    x_values.append(new_x)

    # Check for convergence
    if abs(new_x - x) < tolerance:
        break
    x = new_x
```

```python
# Result
print(f"Local minimum occurs at x = {x:.6f}")
print(f"f(x) at this point is {f(x):.6f}")
```

```python
# Plotting the function and gradient descent steps
x_range = np.linspace(-5, 5, 100)
y_range = f(x_range)

plt.plot(x_range, y_range, label="y = (x + 3)^2", color='blue')
plt.plot(x_values, [f(x) for x in x_values], marker="o", color="red", label="Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to Find Local Minimum")
plt.legend()
plt.grid(True)
plt.show()
```
